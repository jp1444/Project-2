```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
import numpy as np

conn = sqlite3.Connection('database.sqlite')
cur = conn.cursor()
```

```python
cur.execute("""SELECT * FROM Matches WHERE (Season == 2011) AND (Div == 'D1') OR (Season == 2011) AND (Div == 'D2')""")
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
print(df.shape)
df.head()
```

```python
def contains_h(text):
    bool_val = 'H' in text
    return bool_val

def contains_a(text):
    bool_val = 'A' in text
    return bool_val

df['H_W'] = df['FTR'].map(contains_h)
df['A_W'] = df['FTR'].map(contains_a)
```

```python
Wins = pd.DataFrame(df.groupby(['HomeTeam'])['H_W'].sum()+df.groupby(['AwayTeam'])['A_W'].sum())
Losses = pd.DataFrame(df.groupby(['HomeTeam'])['A_W'].sum()+df.groupby(['AwayTeam'])['H_W'].sum())
```

```python
Games = pd.DataFrame(df.groupby(['HomeTeam'])['FTHG'].count()+df.groupby(['HomeTeam'])['FTHG'].count())
```

```python
Goals = pd.DataFrame(df.groupby(['HomeTeam'])['FTHG'].sum()+df.groupby(['AwayTeam'])['FTAG'].sum())
Goals = Goals.rename(columns={0:'Goals'})
```

```python
Goals = Goals.reset_index()
Wins = Wins.reset_index()
Losses = Losses.reset_index()
Games = Games.reset_index()
```

```python
result = Goals.merge(Wins)
result = result.merge(Losses,on='HomeTeam')
result = result.merge(Games,on='HomeTeam')
```

```python
result = result.rename(columns={'HomeTeam': 'Name','0_x':'Wins','0_y':'Losses','FTHG':'Games'})
result['Wins_perc'] = result['Wins']/result['Games']
result['Losses_perc'] = result['Losses']/result['Games']
result.head()
```
